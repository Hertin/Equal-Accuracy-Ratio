```python
import pandas as pd
import os
import json
if 'cd' not in globals():
    os.chdir('..')
    cd = True
```

```python
with open('scripts/labels.json', 'r') as f:
    labels = json.load(f)
def check_characters(string, labels):
    for c in string:
        if c not in labels:
            return False
    return True
```

```python

# data_dir = 'data/primary_microphone/si_tr'
# mode = 'train'

data_dir = 'data/primary_microphone/si_dt'
mode = 'dev'

# data_dir = 'data/primary_microphone/si_et_1'
# mode = 'test'
```

```python

file_trans = []
for speaker_folder in os.listdir(data_dir):
    for filename in os.listdir(f'{data_dir}/{speaker_folder}'):
        
        

        if not filename.endswith('.wrd'):
            continue
        
        if len(file_trans) % 1000 == 0:
            print(f'{len(file_trans)}th file')
#         print(speaker_folder, filename)
        utt, extname = filename.split('.')
        with open(f'{data_dir}/{speaker_folder}/{filename}', 'r') as f:
            trans = []
            for l in f:
                w = l.strip().split()[-1]
                trans.append(w.upper())
            trans = ' '.join(trans)
        if check_characters(trans, labels) and len(trans) > 10:
            file_trans.append((f'{data_dir}/{speaker_folder}/{utt}.wav', trans, speaker_folder))
```
